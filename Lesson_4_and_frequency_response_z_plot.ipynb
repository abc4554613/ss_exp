```python
#(4) 設計一即時FIR系統，使得該系統能完整濾除  𝜔̂ =𝜋7  ,  𝜔̂ =𝜋8  ,  𝜔̂ =𝜋9  之數位信號，繪製其pole zero plot
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
angle = np.linspace(-np.pi, np.pi, 50)
cirx = np.sin(angle)
ciry = np.cos(angle)

a=[1,-np.exp(-1j*np.pi/7)]
z=np.roots(a)

b=[1,-np.exp(-1j*np.pi/8)]
z2=np.roots(b)

c=[1,-np.exp(-1j*np.pi/9)]
z3=np.roots(c)

a1=[1,-np.exp(1j*np.pi/7)]
z4=np.roots(a1)

b1=[1,-np.exp(1j*np.pi/8)]
z5=np.roots(b1)

c1=[1,-np.exp(1j*np.pi/9)]
z6=np.roots(c1)


plt.figure(figsize=(8,8))
plt.plot(cirx, ciry,'k-')
plt.plot(np.real(z), np.imag(z), 'o', markersize=12)
plt.plot(np.real(z6), np.imag(z6), 'o', markersize=12)
plt.plot(np.real(z2), np.imag(z2), 'o', markersize=12)
plt.plot(np.real(z3), np.imag(z3), 'o', markersize=12)
plt.plot(np.real(z4), np.imag(z4), 'o', markersize=12)
plt.plot(np.real(h), np.imag(z5), 'o', markersize=12)

plt.grid()

plt.xlim((-2, 2))
plt.xlabel('Real')
plt.ylim((-2, 2))
plt.ylabel('Imag')
```
