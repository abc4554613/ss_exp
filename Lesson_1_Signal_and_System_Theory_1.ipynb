```python
#coding: utf-8
%matplotlib inline
import numpy as np
import imageio
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed
from IPython.display import display

lena_img = imageio.imread('lena_std.jpg')

lims = (0.1, 1.0, 0.1)

def edit_image(r=1.0, g=1.0, b=1.0):
    lena_img2 = lena_img*0
    lena_img2[:, :, 0] = r*lena_img[:, :, 0]
    lena_img2[:, :, 1] = g*lena_img[:, :, 1]
    lena_img2[:, :, 2] = b*lena_img[:, :, 2]
    
    plt.imshow(lena_img2)

w = interactive(edit_image, r=lims, g=lims, b=lims)
display(w)
```

問題：上網搜尋 skimage.transform.rotate 的使用方法，將上述介面增加一項新的滑桿。
用這個滑桿可以調整Lena影像的旋轉角度。


```python
#coding: utf-8
%matplotlib inline
import numpy as np
from skimage import data
from skimage.transform import rotate
import imageio
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed
from IPython.display import display

lena_img = imageio.imread('lena_std.jpg')

lims = (0.1, 1.0, 0.1)
lim=(0,360,30)
def edit_image(r=1.0, g=1.0, b=1.0,t=360):
    lena_img2 = lena_img*0
    lena_img2[:, :, 0] = r*lena_img[:, :, 0]
    lena_img2[:, :, 1] = g*lena_img[:, :, 1]
    lena_img2[:, :, 2] = b*lena_img[:, :, 2]
    lena_img2=rotate(lena_img2,t)
    
    plt.imshow(lena_img2)

w = interactive(edit_image, r=lims, g=lims, b=lims,t=lim)
display(w)
```

## 問題八: 將兩個旋波相加後，並畫在一起。
    
### (1) 產生兩個cos弦波 
<BR>X1: Amplitude: 1, Phase: 0 rad, Frequency: 5 Hz
<BR>X2: Amplitude: 0.7, Phase: pi rad, Frequency: 5 Hz
<BR>X3 = X1 + X2

### (2)就你的觀察X3的頻率為何？

### (3) 使用np.real, np.exp函式，將cos 表示成 complex exponetial 之實部函數，重覆上述動作

### $$\cos(\omega t) = \mathrm{Re}[ e^{j\omega t}] $$



```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0, 1, 0.001)
x1 = np.cos(2*np.pi*5*t)
x2 = 0.7*np.cos(2*np.pi*5*t+np.pi)
x3=x1+x2
plt.figure(figsize=[10, 5], dpi=100)
plt.plot(t, x1)
plt.plot(t, x2)
plt.plot(t, x3)
plt.grid(True)
```

(2)和x1 x2 頻率依樣


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0, 1, 0.001)
x1 = np.real(np.exp(2*np.pi*5*t*1j))
x2 = 0.7*np.real(np.exp(2*np.pi*5*t*1j+np.pi*1j))
x3=x1+x2
plt.figure(figsize=[10, 5], dpi=100)
plt.plot(t, x1)
plt.plot(t, x2)
plt.plot(t, x3)
plt.grid(True)
```
