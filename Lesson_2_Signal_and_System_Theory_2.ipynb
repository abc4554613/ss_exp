### 傅利葉轉換
### 執行以下程式碼並觀察輸出

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fs = 100
t = np.arange(0, 1, 1/fs)
x = np.cos(2*np.pi*10*t)
plt.figure(figsize=[10,5])
plt.plot(t, x)
plt.title('Signal plot')

xf = np.fft.fft(x)
w_hat = np.arange(0, 2*np.pi, np.pi*2/fs)

plt.figure(figsize=[15,5])

plt.subplot(121)
plt.plot(w_hat, abs(xf))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The magnitude spectrum')

plt.subplot(122)
plt.plot(np.angle(xf))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The phase spectrum')
```

```python
#(1) 下圖左為將信號取快速傅利葉轉換之後的結果，橫軸以數位頻率  𝜔̂   表示，請利用數位頻率與類比頻率的轉換關係，將下圖的橫軸以類比頻率來表示。

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fs = 100
t = np.arange(0, 1, 1/fs)
x = np.cos(2*np.pi*10*t)
plt.figure(figsize=[10,5])
plt.plot(t, x)
plt.title('Signal plot')

xf = np.fft.fft(x)
w_hat = np.arange(0, 2*np.pi, np.pi*2/fs)
f=w_hat*fs/(2*np.pi)
plt.figure(figsize=[15,5])

plt.subplot(121)
plt.plot(f, abs(xf))
plt.xlabel(r'f (hz)')
plt.title('The magnitude spectrum')

plt.subplot(122)
plt.plot(np.angle(xf))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The phase spectrum')

print( np.argmax(abs(xf)))
```

```python
#(2) 使用取最大值 np.max, np.argmax，來自動從頻譜中找出類比頻率
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fs = 100
t = np.arange(0, 1, 1/fs)
x = np.cos(2*np.pi*10*t)


xf = np.fft.fft(x)
w_hat = np.arange(0, 2*np.pi, np.pi*2/fs)
f=w_hat*fs/(2*np.pi)
plt.figure(figsize=[15,5])

plt.subplot(121)
plt.plot(f[0:50], abs(xf)[0:50])
plt.xlabel(r'f (hz)')
plt.title('The magnitude spectrum')
print("最大值的類比頻率")
print (np.argmax(abs(xf)))
```

```python
#(3) 複製以上程式，將弦波頻率改為30，並繪出頻譜，且利用上一題的結果來自動找尋頻譜中的最大peak
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fs = 100
t = np.arange(0, 1, 1/fs)
x = np.cos(2*np.pi*30*t)
plt.figure(figsize=[10,5])
plt.plot(t, x)
plt.title('Signal plot')

xf = np.fft.fft(x)
w_hat = np.arange(0, 2*np.pi, np.pi*2/fs)
f=w_hat*fs/(2*np.pi)
plt.figure(figsize=[15,5])

plt.subplot(121)
plt.plot(f, abs(xf))
plt.xlabel(r'f (hz)')
plt.title('The magnitude spectrum')

plt.subplot(122)
plt.plot(np.angle(xf))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The phase spectrum')
print("最大值")
print(np.max(xf))
print("最大值的頻率")
print( np.argmax(abs(xf)))
```

```python
#(4) 複製以上程式，將弦波頻率改為60，並繪出頻譜，且利用上一題的結果來自動找尋頻譜中的最大peak
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fs = 100
t = np.arange(0, 1, 1/fs)
x = np.cos(2*np.pi*60*t)
plt.figure(figsize=[10,5])
plt.plot(t, x)
plt.title('Signal plot')

xf = np.fft.fft(x)
w_hat = np.arange(0, 2*np.pi, np.pi*2/fs)
f=w_hat*fs/(2*np.pi)
plt.figure(figsize=[15,5])

plt.subplot(121)
plt.plot(f, abs(xf))
plt.xlabel(r'f (hz)')
plt.title('The magnitude spectrum')

plt.subplot(122)
plt.plot(np.angle(xf))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The phase spectrum')
print("最大值")
print(np.max(xf))
print("最大值的頻率")
print( np.argmax(abs(xf)))
```

```python
#(5) 利用上星期實驗七的介面範例，製作兩個滑桿，動態調控弦波頻率以及取樣頻率，並即時以頻譜來獲得最大peak對應之頻率
%matplotlib inline
from ipywidgets import interact, interactive, fixed
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
lims = (0, 100, 10)
lim = (0, 1000, 10)
def freq(f=1, fs=1):
 
    t = np.arange(0, 1, 1/fs)
    x = np.cos(2*np.pi*f*t)
    plt.figure(figsize=[10,5])
    plt.plot(t, x)
    plt.title('Signal plot')

    xf = np.fft.fft(x)
    w_hat = np.arange(0, 2*np.pi, np.pi*2/fs)
    f=w_hat*fs/(2*np.pi)
    plt.figure(figsize=[15,5])

    plt.subplot(121)
    plt.plot(f, abs(xf))
    plt.xlabel(r'f (hz)')
    plt.title('The magnitude spectrum')

    plt.subplot(122)
    plt.plot(np.angle(xf))
    plt.xlabel(r'$\hat\omega$  (radian)')
    plt.title('The phase spectrum')
w = interactive(freq,fs=lim, f=lims )
display(w)
```

```python
#(1) 以上圖x為輸入信號，設計一個有限脈衝響應濾波器 FIR filter, bk = [1/3, 1/3, 1/3]，意即 y[n]= 1/3*x[n] + 1/3*x[n-1] + 1/3*x[n-2]，並畫出輸出信號y[n]
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x=np.array([0.,0,0,2,4,6,4,2,0,0,0])
y=x*0
y[0] = 1/3*(x[0])
y[1] = 1/3*(x[1] + x[0])
for i in range(2,len(x)):
    y[i]=1/3*(x[i]+x[i-1]+x[i-2])




plt.stem(y, )
```
